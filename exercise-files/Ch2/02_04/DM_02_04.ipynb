# Import libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

# Load data

```python
data = pd.read_csv('~/Documents/GitHub/linkedin-learning-data-science-foundations-data-mining/exercise-files/Ch2/02_04/data/b5.csv')
```

## Convert data to numpy array

```python
X = data.values
```

## Scale the values

```python
X = scale(X)
```

# Conduct PCA
## PCA with 50 components

```python
pca = PCA()
pca.fit(X)
print(pca.explained_variance_ratio_)
```

## PCA with 5 components

```python
pca5 = PCA(n_components=5)
pca5.fit(X)
print(pca5.explained_variance_ratio_)
```

## Contributions of variables to components

```python
print(pca5.components_)
```
